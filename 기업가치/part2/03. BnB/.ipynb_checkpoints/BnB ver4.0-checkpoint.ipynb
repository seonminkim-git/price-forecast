# fucntion


```python
def MAPE(y_true, y_hat):
    return np.mean(np.abs((y_hat - y_true) / y_true)) * 100

def RMSE(y_true, y_hat):
    return np.sqrt(np.mean((y_hat - y_true)**2))

def adj_R2(y_true, y_hat, k, n=None):
    rs = r2_score(y_true, y_hat)
    if n is None:  # 소희M test adj_R2 의 n=#train
        n = len(y_true)
    return 1-(1-rs)*(n-1)/(n-k-1)
```

```python
def checkIfDuplicates_1(listOfElems):
    ''' Check if given list contains any duplicates '''
    if len(listOfElems) == len(set(listOfElems)):
        return False
    else:
        return True
    
def flatten(xss):
    return [x for xs in xss for x in xs]
```

```python

def cross_myBnB(combi_num, data, target):
#     target = 'PBR'

    R2, adjR2, R2_train, adjR2_train = [], [], [], []

    X = data.drop(['Date_x', 'close', 'close_weighted', 'EV_EBITDA', 'PBR'], axis='columns')
    if target == 'EV_EBITDA':
        X.drop(['EBITDA2(천원)'], axis='columns', inplace=True)
    elif target == 'PBR':
        X.drop(['BPS(지배, Adj.)(원/주)'], axis='columns', inplace=True)
    y = data[target]
#     X = X.iloc[:,:6]
    # combination list

    combi = list(combinations(X.columns, combi_num)) # Combination
    combi = [i for i in combi if checkIfDuplicates_1(i)==False] # 조합 내 중복 제거
    combi = np.unique(combi,axis=0).tolist() # 전체 조합 중복 제거 

    print(f'combi number: {len(combi)}')

    for i in range(len(combi)) :

        r2, r2_adj = [], []
        r2_train, r2_adj_train = [], []

        cv_k = 5
        cv_rs = 123

        '''split train/test'''
        kf = KFold(n_splits=cv_k, random_state=cv_rs, shuffle=True)

        rmse, mape = [], []
        
        print('* Cross validation *')

        for train_idx, test_idx in kf.split(data):

            train_x = data.iloc[train_idx][list(combi[i])]
            train_y = data.iloc[train_idx][target]


            test_x = data.iloc[test_idx][list(combi[i])]
            test_y = data.iloc[test_idx][target]



            print(f'  {len(rmse)+1} fold ...')

            model = LinearRegression().fit(train_x, train_y)

            '''prediction'''
            pred = model.predict(test_x)
            pred_train = model.predict(train_x)

            '''Scoring'''
            mape += [[MAPE(train_y, pred_train), MAPE(test_y, pred)]]
            rmse += [[RMSE(train_y, pred_train), RMSE(test_y, pred)]]

            r2 += [ r2_score(test_y, pred)]
            r2_adj += [adj_R2(test_y, pred, train_x.shape[1])]

            r2_train += [r2_score(train_y, pred_train)]
            r2_adj_train += [adj_R2(train_y, pred_train, train_x.shape[1])]

            ''' performance '''
#             if len(r2)%100==0:
                
            print('Train performance')
            print('  R^2:', np.round(r2_score(train_y, pred_train), 3))
            print('  adj R^2:', np.round(adj_R2(train_y, pred_train, train_x.shape[1]), 3))
            print('  MAPE:', np.round(MAPE(train_y, pred_train), 3))
            print('  RMSE:', np.round(RMSE(train_y, pred_train), 3))
            print()
            print('Test performance')
            print('  R^2:', np.round(r2_score(test_y, pred), 3))
            print('  adj R^2:', np.round(adj_R2(test_y, pred, test_x.shape[1], n=test_x.shape[0]), 3))
            print('  MAPE:', np.round(MAPE(test_y, pred), 3))
            print('  RMSE:', np.round(RMSE(test_y, pred), 3))

        mape = np.array(mape).T
        rmse = np.array(rmse).T

        R2.append(np.mean(r2))
        adjR2.append(np.mean(r2_adj))

        R2_train.append(np.mean(r2_train))
        adjR2_train.append(np.mean(r2_adj_train))

#     mape = [np.mean(a) for a in mape]
#     rmse = [np.mean(a) for a in rmse]
#     print('  -> MAPE (train/valid):', ' / '.join([str(n) for n in np.round(mape, 1)]))
#     print('  -> RMSE (train/valid):', ' / '.join([str(n) for n in np.round(rmse, 1)]))
#     print()


    result = pd.DataFrame(combi)
    result['R2'] = R2
    result['adjR2'] = adjR2
    result['R2_train'] = R2_train
    result['adjR2_train'] = adjR2_train
    result['mean_adj_R2_train_test'] = (np.array(adjR2)+np.array(adjR2_train))/2  #seonmin_rev



    return result
```

# crossBnB

```python
# file_list =['/Users/soheehwang/Documents/013.2022_SK_기업주가분석2/data/ITS_v1.3/Accenture_non_shift_clean_v1.3.csv',
#  '/Users/soheehwang/Documents/013.2022_SK_기업주가분석2/data/ITS_v1.3/Infosys_non_shift_clean_v1.3.csv',
#  '/Users/soheehwang/Documents/013.2022_SK_기업주가분석2/data/ITS_v1.3/Tcs_non_shift_clean_v1.3.csv']
file_list =[os.path.join(datafile_path, "Accenture_non_shift_clean_v1.3.csv"),
            os.path.join(datafile_path, "Infosys_non_shift_clean_v1.3.csv"),
            os.path.join(datafile_path, "Tcs_non_shift_clean_v1.4.csv")]
file_list
```

```python
top_total = []
frequece_total = []
company_name_total = []
y_total = []

''' Setting '''
combi_itter = 2
log_type = 'normal'
# save_dir = '/Users/soheehwang/Documents/013.2022_SK_기업주가분석2/output/BNB6/'
save_dir = os.path.join(current_dir, "output", "BNB6")


if not os.path.exists(save_dir):
    os.makedirs(save_dir)
    
for file_path in file_list:

    folder_name = file_path.split('\\')[-2]
    file_name = file_path.split('\\')[-1].split('.')[0]
    company_name = file_name.split('_')[0]

    for target in ['EV_EBITDA']:

        ''' Setting: 필수 포함 cols '''
        
        core_ev = ['매출액증가율(YoY)(연도)', 'Gross Investment (GI)', '투자비 대비 매출 (Rev/GI)s', '무형자산회전율s','매출총이익(천원)']

        start = time.time()
        start_time = str(datetime.timedelta(seconds=start)).split(".")
        print(f'{company_name}_{target}: Start. {start_time}')
        
        ''' data load '''
        
        data = pd.read_csv(f"{file_path}")        
        display(data.head(5))
        data = data.sort_values('Date_x').reset_index(drop=True) # sort by date
        print(f'data shape: {data.shape}')
        
        ''' log 변환 '''
        origin = data.columns[:-1].to_list()
        
        ## logx 변환
        if log_type == 'logx':
            data[origin] = np.log(data[origin])
        
        ## logy 변환
        elif log_type == 'logy':
            data[data.columns[-1]] = np.log(data[data.columns[-1]])
        
        ## 전체 log 변환
        elif log_type == 'whole':
            data = np.log(data)
        
        data = data.replace([np.inf, -np.inf], np.nan)
        data = data.fillna(method='bfill').fillna(method='ffill').dropna(axis=1)
        core_ev.extend(['Date_x', 'close', 'close_weighted', 'EV_EBITDA', 'PBR', 'EBITDA2(천원)'])
        
        data = data[core_ev]        
        print(data.shape)
 
        temp = cross_myBnB(combi_num=combi_itter, data=data, target=target)
        

        bnb_final = temp.sort_values(['mean_adj_R2_train_test'], ascending = False)  #seonmin_rev
        
        top = bnb_final.head(100).copy()
        display(bnb_final.head())

        ''' frequency 및 rank 계산 '''
        total = []

        for _ in range(combi_itter):

            total.append(list(top[_]))
        total = flatten(total)

        my_dict = Counter(total)
        sort_df = pd.DataFrame.from_dict(my_dict, orient='index').reset_index()
        sort_df = sort_df.rename(columns={0: 'freq'})
        sort_df['rank'] = sort_df['freq'].rank(method='min',ascending=False)
        sort_df.sort_values('rank', inplace=True)
        sort_df.rename(columns={0: 'freq'})
        
        ''' 중간 데이터 저장 '''
        
        if not os.path.exists(save_dir+f'/{folder_name}'):
            os.makedirs(save_dir+f'/{folder_name}')
        top.to_csv(f'{save_dir}/{folder_name}/{company_name}_{target}_{combi_itter}_{log_type}_top100.csv')
        sort_df.to_csv(f'{save_dir}/{folder_name}/{company_name}_{target}_{combi_itter}_{log_type}_frequence.csv')
        print(f'{company_name}_{target}: Save Complete.')

        top_total.append(top)
        frequece_total.append(sort_df)
        company_name_total.append(company_name)
        y_total.append(target)
        
        end = time.time()
        sec = (end - start)
        result_list = str(datetime.timedelta(seconds=sec)).split(".")
        print(result_list[0])

        
''' 전체 결과 취합 후 저장 '''

master_EV = master[['variable', 'group', 'importance']]

for _ in range(len(top_total)):

    sub = frequece_total[_].set_index('index')[['freq', 'rank']]
    sub= sub.rename(columns={'rank': f'{company_name_total[_]}_rank', 'freq': f'{company_name_total[_]}_freq'})
    master_EV = master_EV.join(sub, on='variable', how='left', lsuffix='', rsuffix='', sort=False)
    display(master_EV.head(5))
    final_evitda = master_EV[master_EV.group.isin(['성장성', '수익성','안정성','활동성'])].fillna(0)
    final_evitda.to_excel(f'{save_dir}/{folder_name}/{y_total[_]}_{combi_itter}_{log_type}.xlsx')
        
# 1. 엑셀 파일 열기 w/ExcelWriter
writer = pd.ExcelWriter(f'{save_dir}/{folder_name}/ITSgroup_freq_summary_{log_type}.xlsx', engine='xlsxwriter')

# 2. 시트별 데이터 추가하기(1개 시트로 저장할때와 유사)
final_evitda.to_excel(writer, sheet_name= f'{y_total[1]}_freq_rank_{log_type}')

# 3. 엑셀 파일 저장하기
writer.save()
```
